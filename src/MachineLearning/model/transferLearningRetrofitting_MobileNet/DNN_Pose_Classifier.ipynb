### data resizing and classification

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator as dataGen

#lets target the scaling res
trainDatagen = dataGen(rescale=1/255) #1/255
validationDatagen = dataGen(rescale=1/255) #1/255
"""
Chair
Cobra
Dog
Tree
Warrior
"""
trainDataGenerator = trainDatagen.flow_from_directory(
    trainingPath,
    classes = ['chair', 'cobra', 'dog', 'tree', 'warrior'],
    target_size = (300, 300),
    batch_size = 32,
    class_mode='categorical',
)

trainingValidationGenerator = validationDatagen.flow_from_directory(
    validationPath,
    classes = ['chair', 'cobra', 'dog', 'tree', 'warrior'],
    target_size = (300, 300),
    batch_size = 8,
    class_mode='categorical',
    shuffle=False
)

#trainDataGenerator = np.asarray(trainDataGenerator).astype('float32').reshape((-1,1))
#trainingValidationGenerator = np.asarray(trainingValidationGenerator).astype('float32').reshape((-1,1))
```

# Train time!

```python
ACCURACY_THRESHOLD = 0.91
class leCallback(tf.keras.callbacks.Callback): 
    def on_epoch_end(self, epoch, logs={}): 
        if(logs.get('accuracy') > ACCURACY_THRESHOLD):   
          print("\nReached %2.2f%% accuracy, so stopping training!!" %(ACCURACY_THRESHOLD*100))  
          tf.keras.callbacks.ModelCheckpoint('./learningState', save_best_only=True) 
          self.model.stop_training = True
```

```python
savestateCheckPoint = leCallback()
history = model.fit(trainDataGenerator,
steps_per_epoch=8,
epochs=1000,
verbose=1,
validation_data = trainingValidationGenerator,
validation_steps=8,
callbacks=[savestateCheckPoint])
```

### converting h5 to tflite

```python
# Write the classificaiton
! echo "chair" > motionful_prealpha_classification.txt
! echo "cobra" >> motionful_prealpha_classification.txt
! echo "dog" >> motionful_prealpha_classification.txt
! echo "tree" >> motionful_prealpha_classification.txt
! echo "warrior" >> motionful_prealpha_classification.txt
```

```python
modelLoad = tf.keras.models.load_model('motionful_prealpha.h5')
converterModel = tf.lite.TFLiteConverter.from_keras_model(modelLoad)
print('loading converter')
converterModel.optimizations = [tf.lite.Optimize.DEFAULT]
def representative_data_gen():
  for input_value in tf.data.Dataset.from_tensor_slices(trainDataGenerator).batch(1).take(100):
    yield [input_value]
print('loading 2')
converterModel.representative_dataset = representative_data_gen #shows error when value cant be quantized
print('loading 3')
converterModel.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]#force parameter to be uint8 to prevent java buffer error on the android side of things
print('loading 4')
converterModel.inference_input_type = tf.uint8
print('loading 5')
converterModel.inference_output_type = tf.uint8


print('execute!')
tfLiteModelExport = converterModel.convert()
interpreter = tf.lite.Interpreter(model_content=converterModel)
input_type = interpreter.get_input_details()[0]['dtype']
print('input: ', input_type)
output_type = interpreter.get_output_details()[0]['dtype']
print('output: ', output_type)
open("motionful_prealpha.tflite", "wb").write(tfLiteModelExport)
```
